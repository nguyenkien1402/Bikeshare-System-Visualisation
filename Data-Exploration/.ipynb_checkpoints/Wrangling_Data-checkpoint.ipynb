```r
h <- austin_bike_trip
```

```r
h$start_station_latitude <- 0
h$start_station_longitude <- 0
h$end_station_latitude <- 0
h$end_station_longitude <- 0
```

```r
head(austin_station)
```

```r
start_station_lat_f <- function(x){
    start_station_id <- x[[7]]
    start_latitude <- austin_station[austin_station$station_id == start_station_id,1]
    start_latitude
}
start_station_long_f <- function(x){
    start_station_id <- x[[7]]
    start_longitude <- austin_station[austin_station$station_id == start_station_id,3]
    start_longitude
}

stop_station_lat_f <- function(x){
    stop_station_id <- x[[4]]
    stop_latitude <- austin_station[austin_station$station_id == stop_station_id, 1]
    stop_latitude
}

stop_station_long_f <- function(x){
    stop_station_id <- x[[4]]
    stop_longitude <- austin_station[austin_station$station_id == stop_station_id, 3]
    stop_longitude
}
```

```r
nrow(h)
```

```r
h$start_station_latitude = apply(h, 1, start_station_lat_f)
h$start_station_longitude = apply(h, 1, start_station_long_f)
h$end_station_latitude = apply(h, 1, stop_station_lat_f)
h$end_station_longitude = apply(h, 1, stop_station_long_f)
```

All the code above is used to concatenate the longitude and latitude of station into one single dataframe

Eliminate all the start_station_id and end_station_id which have null value

```r
# check if the station_id is null
h <- h %>%
         filter(!is.na(start_station_id),!is.na(end_station_id))
nrow(h)
```

Bike ID does not related to our visualization, so we can get rid of it
Now,we are going to fill down all the null year by using start_time


```r
h$id <- sample(1:nrow(h), replace = FALSE)
h$start_time <- as.character(h$start_time)
```

```r
# fill down the year 

k <- h[is.na(h$year),]
fill_down_year_f <- function(x){
    start_time <- x$start_time
    year <- strsplit(x$start_time,"-")[[1]][1]
    year
}

fill_down_month_f <- function(x){
    start_time <- x$start_time
    month <- strsplit(x$start_time, "-")[[1]][2]
    month
}
```

```r
k$year <- apply(k, 1, fill_down_year_f)
k$month <- apply(k, 1, fill_down_month_f)
```

```r
# h[is.na(h$year),] <- apply(h[is.na(h$year),], 1, fill_down_year_f)
# h[is.na(h$month),] <- apply(h[is.na(h$month),], 1, fill_down_month_f)
h[is.na(h$year),"year"] <- k$year
h[is.na(h$month),"month"] <- k$month
```

```r
# get the same station id and end station id
same_station <- h %>% 
    filter(start_station_id == end_station_id)

#get rid of all the data
h <- h %>% filter(start_station_id != end_station_id)
```

```r
austin_weather <- austin_weather %>% select(Date, TempHighF, TempAvgF, DewPointHighF, DewPointAvgF,
                                            HumidityHighPercent, HumidityAvgPercent,
#                                           SeaLevelPressureHighInches, SeaLevelPressureAvgInches, VisibilityHighMiles,VisibilityAvgMiles,
                                           WindHighMPH, WindAvgMPH, PrecipitationSumInches, Events)
```

```r
# make a date column in austin_bike_trip
h$Date <- 0

fill_down_date_f <- function(x){
    start_time <- x$start_time
    Date <- strsplit(start_time, " ")[[1]][1]
    Date
}

h$Date <- apply(h, 1, fill_down_date_f)
```

```r
# now, going to join data
new_h <- left_join(x = h, y=austin_weather, by="Date")
```

```r
# convert starT_time to factor 
new_h$start_time <- as.factor(new_h$start_time)
# convert year to int
new_h$year <- as.numeric(new_h$year)
# convert month to int
new_h$month <- as.numeric(new_h$month)
```

```r
# now, trim the text and finalize data
new_h$start_station_name <- trimws(new_h$start_station_name)
new_h$end_station_name <- trimws(new_h$end_station_name)
```
